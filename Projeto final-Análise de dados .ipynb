```python
#Importar a biblioteca pandas 
import pandas as pd 
```

```python
#Importar a tabela de vendas dez
vendas_dez= pd.read_excel('Vendas - Dez.xlsx')
display (vendas_dez)
```

```python
vendas_dez.columns
```

```python
vendas_dez.columns= ['Nota Fiscal', 'Data', 'ID Loja', 'Produto', 'Quantidade', 'Valor Unitário', 'Faturamento']
display (vendas_dez)
```

```python
#Análise do faturamento total da empresa
fat_total= vendas_dez ['Faturamento'].sum()
display(fat_total)
```

```python
#Faturamento por loja 
fat_loja= vendas_dez[['ID Loja', 'Faturamento']].groupby('ID Loja'). sum()
display (fat_loja)
```
